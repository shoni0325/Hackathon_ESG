### Logistic Regression

```python
df = pd.read_csv('../../data/new/winsor_1.csv', encoding='euc-kr', index_col='기업명').drop('Unnamed: 0', axis=1)
df['총자산규모'] = np.log(df['총자산규모'])
x = df.iloc[:,2:40]
y = df['배당기업구분']
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0, stratify=y)
lr = LogisticRegression(random_state=0)
lr.fit(x_train, y_train)
Pred = lr.predict(x_test)
accuracy = accuracy_score(y_test, Pred)

print(f'로지스틱 회귀 예측 정확도 : {accuracy:.4f}')
```

```python
from sklearn.model_selection import GridSearchCV

params = {'penalty' : ['l2', 'l1'],
          'C' : [0.01, 0.1, 1, 1, 5, 10]}

grid_clf = GridSearchCV(lr, param_grid=params, scoring='accuracy', cv=3)
grid_clf.fit(x_train, y_train)

print(f'최적 하이퍼 파라미터 : {grid_clf.best_params_}, 최적 평균 정확도 : {grid_clf.best_score_:.3f}')
```
