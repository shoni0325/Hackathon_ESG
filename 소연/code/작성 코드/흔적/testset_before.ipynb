## 21-22 주가 데이터

```python
df_2122 = pd.read_csv('./data/ESGdata/업종추가데이터/df_2122_add_job.csv', index_col='종목명')
mv_2021 = pd.read_csv('./data/ESGdata/주가데이터/mv2021.csv', encoding='euc-kr', index_col='종목명')
mv_2022 = pd.read_csv('./data/ESGdata/주가데이터/mv2022.csv', encoding='euc-kr', index_col='종목명')
```

```python
# 주가데이터 종가만 가져오기
mv_2021 = pd.DataFrame(mv_2021['종가'])
mv_2022 = pd.DataFrame(mv_2022['종가'])
```

```python
# 각 리스트 저장
list_2122 = df_2122.index
list_mv2021 = mv_2021.index
list_mv2022 = mv_2022.index
```

```python
# 21-22 업종데이터와 21년 주가 데이터 공통된 리스트 출력
set1 = set(list_2122)
set2 = set(list_mv2021)

common_value = list(set1 & set2)
print(common_value)
len(common_value)
```

```python
# 공통 리스트 기반 21년 주가데이터 출력
mv_2021 = mv_2021.loc[common_value]
mv_2021.sort_index(inplace=True)
```

```python
# 21-22 업종데이터와 22년 주가데이터 공통된 리스트 출력
set1 = set(list_2122)
set2 = set(list_mv2022)

common_value = list(set1 & set2)
print(common_value)
len(common_value)
```

```python
# 공통 리스트 기반 22년 주가데이터 출력
mv_2022 = mv_2022.loc[common_value]
mv_2022.sort_index(inplace=True)
```

```python
def transform(df):
    
    df.drop('Unnamed: 0', axis=1, inplace=True)
    
    df['업종명'] = df['업종명'].map({
        '건설업': '건설업',
        '기타금융' : '금융 및 보험업',
        '증권': '금융 및 보험업',
        '보험' : '금융 및 보험업',
        '은행' : '금융 및 보험업',
        '서비스업' : '서비스업',
        '유통업' : '운수 및 창고업',
        '운수창고업' : '운수 및 창고업',
        '전기가스업' : '에너지업',
        '통신업' : '정보통신업',
        '섬유의복' : '제조업',
        '음식료품' : '제조업',
        '기타제조업' : '제조업',
        '종이목재' : '제조업',
        '철강금속' : '제조업',
        '기계' : '제조업',
        '운수장비' : '제조업',
        '화학' : '제조업',
        '전기전자' : '제조업',
        '비금속광물' : '제조업',
        '의료정밀' : '제조업',
        '농업, 임업 및 어업' : '1차 산업',
        '광업' : '1차 산업',
        '의약품' : '의약품'
    })


    df_1 = df.set_index(['업종명', '종목명']).sort_index()

    return df_1
  
```

## 이상치 제거

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
origin_2122 = pd.read_csv('./data/origindata/original_2122.csv')

# 데이터 예시
data = origin_2122['등락률']

# 이상치 임계값 설정 (98%)
threshold = np.percentile(data, 98)

# box plot 그리기
plt.boxplot(data)

# 이상치 표시
outliers = [val for val in data if (val > threshold) | (val < -threshold)]
for outlier in outliers:
    plt.plot(1, outlier, 'ro')     # 이상치를 빨간색으로 표시

plt.show()

outliers = [val for val in data if (val > threshold) | (val < -threshold)]
print("이상치 데이터 : ", outliers)
len(outliers)

# 이상치 제외한 값을 원본에 저장
origin_2122 = origin_2122[~origin_2122['등락률'].isin(outliers)]

# origin_2122.to_csv('98%이상치제거_original_2122.csv')
origin_2122
```
