```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../data/new/Test.csv')
```

```python
new_order =['Unnamed: 0','업종명','평가연도','총자산규모(Size)','부채비율(DEB)','자기자본비율(DEB)','매출액증가율','현금흐름(CF)','PBR','ESG','E','S','G','배당률']
df = df.reindex(columns=new_order)
df.columns=['기업명','업종명','평가연도','총자산규모','부채비율','자기자본비율','매출액증가율','현금흐름','PBR','ESG','E','S','G','배당률']
df
```

```python
def change(x):
    if x==0:
        return '무배당기업'
    elif x>=47.12:
        return '고배당기업'
    else:
        return '일반배당기업'
```

```python
df['배당률']=df['배당률'].apply(change)

df=pd.get_dummies(df,columns=['ESG','E','S','G'])
```

```python
df = df[[col for col in df.columns if col != '배당률'] + ['배당률']]
df
```

```python
df['ESG_A+'] = False
df['ESG_B이하'] = False
df['E_B+ 이하'] = False
df['S_B이하'] = False
df['G_-'] = False
df['G_B이하'] = False
df.set_index('기업명')
```

```python
df = df[[col for col in df.columns if col != '배당률'] + ['배당률']]
```
