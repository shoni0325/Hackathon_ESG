```python
import pandas as pd
```

```python
#########
kospi2018=pd.read_csv('./job/2018코스피업종분류.csv',encoding='euc-kr',index_col='종목명')
kospi2017=pd.read_csv('./job/2017코스피업종분류.csv',encoding='euc-kr',index_col='종목명')
kospi2016=pd.read_csv('./job/2016코스피업종분류.csv',encoding='euc-kr',index_col='종목명')
kospi2015=pd.read_csv('./job/2015코스피업종분류.csv',encoding='euc-kr',index_col='종목명')
kospi2014=pd.read_csv('./job/2014코스피업종분류.csv',encoding='euc-kr',index_col='종목명')
kospi2013=pd.read_csv('./job/2013코스피업종분류.csv',encoding='euc-kr',index_col='종목명')
kospi2012=pd.read_csv('./job/2012코스피업종분류.csv',encoding='euc-kr',index_col='종목명')
kospi2011=pd.read_csv('./job/2011코스피업종분류.csv',encoding='euc-kr',index_col='종목명')

##########
kospi2011_list=kospi2011.index.tolist()
kospi2012_list=kospi2012.index.tolist()
kospi2013_list=kospi2013.index.tolist()
kospi2014_list=kospi2014.index.tolist()
kospi2015_list=kospi2015.index.tolist()
kospi2016_list=kospi2016.index.tolist()
kospi2017_list=kospi2017.index.tolist()
kospi2018_list=kospi2018.index.tolist()

##########
df_11=pd.read_csv('./ESG_csv/2011.csv',encoding='euc-kr').fillna('C')
df_12=pd.read_csv('./ESG_csv/2012.csv',encoding='euc-kr').fillna('C')
df_13=pd.read_csv('./ESG_csv/2013.csv',encoding='euc-kr').fillna('C')
df_14=pd.read_csv('./ESG_csv/2014.csv',encoding='euc-kr').fillna('C')
df_15=pd.read_csv('./ESG_csv/2015.csv',encoding='euc-kr').fillna('C')
df_16=pd.read_csv('./ESG_csv/2016.csv',encoding='euc-kr').fillna('C')
df_17=pd.read_csv('./ESG_csv/2017.csv',encoding='euc-kr').fillna('C')
df_18=pd.read_csv('./ESG_csv/2018.csv',encoding='euc-kr').fillna('C')

##########
df_11.set_index('종목명',drop=True,inplace=True)
df_12.set_index('종목명',drop=True,inplace=True)
df_13.set_index('종목명',drop=True,inplace=True)
df_14.set_index('종목명',drop=True,inplace=True)
df_15.set_index('종목명',drop=True,inplace=True)
df_16.set_index('종목명',drop=True,inplace=True)
df_17.set_index('종목명',drop=True,inplace=True)
df_18.set_index('종목명',drop=True,inplace=True)

##########
name_11=df_11.index.to_list()
name_12=df_12.index.to_list()
name_13=df_13.index.to_list()
name_14=df_14.index.to_list()
name_15=df_15.index.to_list()
name_16=df_16.index.to_list()
name_17=df_17.index.to_list()
name_18=df_18.index.to_list()
```

## 2011

```python
set1=set(kospi2011_list)
set2=set(name_11)
common_values = list(set1 & set2)

list1=df_11.loc[common_values]
list1=list1.loc[~list1.index.duplicated()]
list1.sort_index(inplace=True)

list2=kospi2011.loc[common_values].sort_index()[['업종명']]
df_11_add_job=pd.concat([list1,list2],axis=1)
df_11_add_job
```
