## Diff 21/22
* 739개에서 691개로 감소

```python
df_21 = pd.read_csv('./data/ESGdata/개별ESG데이터/2021.csv', encoding='euc-kr', index_col=0).drop(['ESG 통합등급', '평가연도'], axis=1)
df_22 = pd.read_csv('./data/ESGdata/개별ESG데이터/2022.csv', encoding='euc-kr', index_col=0).drop(['ESG 통합등급', '평가연도'], axis=1)
```

```python
diff_2122=pd.merge(df_21,df_22,how='inner',on='종목명')
diff_2122['Diff_E']=''
diff_2122['Diff_S']=''
diff_2122['Diff_G']=''
diff_2122.set_index('종목명',drop=True,inplace=True)
diff_2122
```

## diff_2122 파일에 업종 추가
* 사전 작업
    * 위에서 만든 코스피업종분류 -> 코스피 종목명 리스트
        * 코스피 + 코스닥 합쳐있는 자료 나오면 해당 파일로 다시 종목명 리스트 추출
    * 감소한 데이터 수 확인
    * 코스닥, 코스피 데이터로 원래 데이터 종목명에 set()

```python
name_2122 = diff_2122.index.tolist()
```

```python
set1 = set(industry_2022_list)  # 더 많아서 이걸로 set
set2 = set(name_2122)

common_value = list(set1&set2)
print(common_value)
len(common_value)
```

```python
industry_2022.loc[common_value].sort_index()[['업종명']]
```

```python
list1 = diff_2122.loc[common_value]
list1 = list1.loc[~list1.index.duplicated()]
list1.sort_index(inplace=True)
```

```python
list2 = industry_2022.loc[common_value].sort_index()[['업종명']]
list2
```

```python
df_2122_add_job = pd.concat([list1, list2], axis=1)
df_2122_add_job
#df_2122_add_job.to_csv('df_2122_add_job.csv')
```
